```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as mp
```

```python
tips=sns.load_dataset('tips')
iris=sns.load_dataset('iris')
iris.head()
tips.head()
```

```python
sns.histplot(tips['total_bill'],kde=False)
plt.title("histogram")
plt.show()
```

```python
# KDE Plot
sns.kdeplot(tips['total_bill'])
plt.title('KDE Plot - Smooth Distribution of Total Bill')
plt.show()
```

```python
# Displot (Histogram + KDE)
sns.displot(tips['total_bill'], kde=True)
plt.title('Displot - Total Bill Distribution')
plt.show()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
plt.show()
```

```python
# Count Plot - Frequency of Days
sns.countplot(x='day', data=tips)
plt.title('Count Plot - Frequency of Days')
plt.show()
```

```python
# Box Plot - Total Bill by Day
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot - Total Bill by Day')
plt.show()
```

```python
# Violin Plot - Distribution + Summary
sns.violinplot(x='day', y='total_bill', data=tips)
plt.title('Violin Plot - Total Bill by Day')
plt.show()
```

```python
# Scatter Plot - Total Bill vs Tip
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='sex')
plt.title('Scatter Plot - Total Bill vs Tip')
plt.show()
```

```python
# Line Plot - Tip vs Party Size (useful for trend-like visuals)
sns.lineplot(x='size', y='tip', data=tips,hue='sex')
plt.title('Line Plot - Tip vs Party Size')
plt.show()
```

```python
# Relplot with Facets
sns.relplot(x='total_bill', y='tip', hue='sex', col='day', data=tips)
plt.suptitle('Relplot - Tip vs Total Bill by Day & Gender')
plt.show()
```

```python
# Select only numeric columns
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation
corr = iris_num.corr()

# Plot the heatmap
sns.heatmap(corr, annot=True,cmap='coolwarm')
plt.title("Heatmap - Correlation of Iris Features")
plt.show()
```

```python
# Select only numeric columns (exclude 'species')
iris_num = iris.select_dtypes(include=['float', 'int'])

# Compute correlation matrix
corr = iris_num.corr()

# Create the clustermap
sns.clustermap(corr, cmap='coolwarm', annot=True)
plt.suptitle('Clustermap - Grouped Correlations (Numeric Only)',y=1.02)
plt.show()
```

```python
mode=df['Customer Remarks'].mode()
df['Customer Remarks'].fillna(mode,inplace=True)
df.isnull().sum()
```

```python
# Handle missing values

#Numeric columns  fill with mean
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
for col in numeric_cols:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)

# Object (string/categorical) columns  fill with mode (most frequent value)
object_cols = df.select_dtypes(include=['object']).columns
for col in object_cols:
    mode_value = df[col].mode()[0] if not df[col].mode().empty else "Unknown"
    df[col].fillna(mode_value, inplace=True)

#Check if any missing values remain
print("Missing values after imputation:")
print(df.isnull().sum())
```

```python
#A.1.Numerical columns
sns.histplot(df['Item_price'],color='blue')
plt.title("histogram-Itemprice")
plt.show()
```

```python
#A.2.CSAT
plt.boxplot(x='CSAT Score',data=df)
plt.title("Boxplot-CSAT")
plt.show()
```

```python
sns.violinplot(df['connected_handling_time'])
plt.title("connected handling time")
plt.show()
```
